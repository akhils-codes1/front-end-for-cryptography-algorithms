```python
def split_into_bytes(s):
    #s = bitstring
    res = []
    j = 0
    while(j<len(s)):
        try:
            res+=[s[j:j+8]]
        except IndexError:
            res+=[s[j:].zfill(8)]
        j = j+8
    return res
```
